This is a Jupyter notebook to visualze a function of two variables, $ z (x,y) $.

It was written and tested using Python 3.7.6 in August, 2021 by Mike Sokoloff.

The purpose is to make a 3-dimensional plot of $ z(x,y) $ to help create a better intuitive feeling for what gradients of functions of two variables as a step towards developing a better understanding of gradients of functions of three variables.

In the cell below, the code imports libraries that will be used for doing calculations and making plots.

[numpy.arange](https://numpy.org/doc/stable/reference/generated/numpy.arange.html) creates a list of evenly spaced values

[math](https://docs.python.org/3/library/math.html) provides access to constants and mathematical functions

[matplotlib.pyplot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.html) is the workhorse for plotting in scientific Python

[pylab](https://matplotlib.org/2.0.2/faq/usage_faq.html#matplotlib-pyplot-and-pylab-how-are-they-related) pylab is a convenience module that bulk imports matplotlib.pyplot (for plotting) and numpy (for mathematics and working with arrays) in a single name space. Although many examples use pylab, it is no longer recommended. But it does what we want here. *Caveat Emptor*

[mpl_toolkits.mplot3d](https://matplotlib.org/2.2.2/mpl_toolkits/index.html)  provides some basic 3D plotting


```python
import numpy as np
import math
import matplotlib.pyplot as plt
from pylab import meshgrid,cm,imshow,contour,clabel,colorbar,axis,title,show

from mpl_toolkits.mplot3d import Axes3D
```

```python
## here, we define a quadratic function to be evaluated for a range of (x,y)

def z_func(x,y):
 return ( (x*x - 2*y*y - 3*x*y - 3*x +5*y -4))
```

```python
## the np.arange commands define numpy arrays spanning the indicated ranges
## if y ou are not familiar with this command, add
## print(' x = ',x) and print(' y = ',y) commands after y = np.arange(-5,5,0.1)

x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)

## the meshgrid command creates a matrix and returns the x & y coordinates of each
## point in the matrix in the 2-D arrays X & Y

X,Y = meshgrid(x, y) # grid of point

## Z is a 2-D array with the values of z_func calculated at each point defined
## by X & Y
Z = z_func(X,Y)

## matplot.pyplot magic -- ask Prof. Google about the details
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                      cmap=cm.RdBu,linewidth=0, antialiased=False)
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
## this cell does essentially the same work as that above, but
## the (x,y) range is "zoomed-in"

x = np.arange(-0,3,0.1)
y = np.arange(-0.5,0.5,0.05)
#print("x = ",x)
print("len(x) = ",len(x))
X,Y = meshgrid(x, y) # grid of point
Z = z_func(X,Y)

fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                      cmap=cm.RdBu,linewidth=0, antialiased=False)

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

The following cell does exactly the same work as the cell above, 
but uses a different color map (cmap=cm.inferno rather than cmap=cm.RdBu)

many[colormaps](https://matplotlib.org/stable/tutorials/colors/colormaps.html) 
are pre-defined in matplotlib; choose one that conveys the ideas well, or create
your own.

```python
x = np.arange(-0,3,0.1)
y = np.arange(-0.5,0.5,0.05)
#print("x = ",x)
print("len(x) = ",len(x))
X,Y = meshgrid(x, y) # grid of point
Z = z_func(X,Y)

fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                      cmap=cm.inferno,linewidth=0, antialiased=False)

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
